```python
#BIBLIOTECAS IMPORTANTES
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn import svm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carregamento da base de dados e criação do dataframe
wine = load_wine()
data = pd.DataFrame(data=wine['data'], columns=wine['feature_names'])
data['target'] = wine['target']
```

```python
i1 = np.random.permutation(data.shape[0])
d1 = int(0.4*len(i1)) # 40% dos termos

train1, test1 = i1[:d1], i1[d1:]

ctrain1, ctest1 = data.loc[train1,:], data.loc[test1,:]

xtrain1 = ctrain1.drop('target', axis=1)
ytrain1 = ctrain1.target

xtest1 = ctest1.drop('target', axis=1)
ytest1 = ctest1.target

clf = svm.SVC(C = 1.0)

clf.fit(xtrain1, ytrain1)

y_pred = clf.predict(xtest1)

accuracy = accuracy_score(ytest1, y_pred)
accuracy

cm1 = confusion_matrix(ytest1, y_pred)
#Matriz de confusão, seguindo os padrões já desenvolvidos

plt.figure(figsize=(8, 6))
sns.heatmap(cm1, annot=True, fmt='d', cmap='Blues', annot_kws={"size": 14})
plt.xlabel('Previsões')
plt.ylabel('Valores Verdadeiros')
plt.title('Matriz de Confusão')
plt.show()
```
